## Cargar los archivos y visualizar
### Infraestructura ciclista
#### Ciclovias

```python
ciclovia = gpd.read_file(ruta_shp+ "infraestructura_vial_ciclista/Ciclovia.shp")
```

```python
ciclovia.head()
```

```python
print(ciclovia.crs)
```

```python
ciclovia.explore()
```

```python
ciclovia_limpia = ciclovia[ciclovia["ESTADO"] == "En operacion"][
    ["ALCALDIA", "TIPO_IC", "TIPO_VIA", "TIPO_CONEC", "LONG_KM", "geometry"]
]
```

```python
ciclovia_limpia.head()
```

```python
ciclovia_limpia = ciclovia_limpia.to_crs(epsg=4326)
```

```python
ciclovia_limpia.explore()
```

```python
ciclovia_limpia.to_file("../datos/procesados/infra_ciclista_limpia/ciclovia_limpia.gpkg", driver="GPKG")
```
