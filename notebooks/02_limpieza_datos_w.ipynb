## Cargar los archivos y visualizar
### 1. Infraestructura ciclista
#### Ciclovias

```python
ciclovia = gpd.read_file(ruta_shp+ "infraestructura_vial_ciclista/Ciclovia.shp")
```

```python
ciclovia.head()
```

```python
print(ciclovia.crs)
```

```python
ciclovia.explore()
```

```python
ciclovia_limpia = ciclovia[ciclovia["ESTADO"] == "En operacion"][
    ["ALCALDIA", "TIPO_IC", "TIPO_VIA", "TIPO_CONEC", "LONG_KM", "geometry"]
]
```

```python
ciclovia_limpia.head()
```

```python
ciclovia_limpia = ciclovia_limpia.to_crs(epsg=4326)
```

```python
ciclovia_limpia.explore()
```

```python
ciclovia_limpia.to_file("../datos/procesados/infra_ciclista_limpia/ciclovia_limpia.gpkg", driver="GPKG")
```

## 4. Rutas RTP

```python
rtp_lineas = gpd.read_file( ruta_shp+ "rtp_shp/RTP_lineas.shp")
rtp_lineas.head()
```

```python
rtp_paradas = gpd.read_file(ruta_shp + "rtp_shp/RTP_paradas.shp")
rtp_paradas.head()
```

```python
rtp_lineas_limpia = rtp_lineas[rtp_lineas["ESTATUS"] == "ACTIVA"][
    ["NOMBRE", "SISTEMA", "RUTA", "ORIGEN", "DESTINO", "geometry"]
]
rtp_paradas_limpia = rtp_paradas[
    ["RUTA", "SENTIDO", "ORIG_DEST", "SISTEMA", "geometry"]
]
```

```python
rtp_lineas_limpia.to_file("../datos/procesados/transporte/rtp_lineas.gpkg", driver="GPKG")
rtp_paradas_limpia.to_file("../datos/procesados/transporte/rtp_paradas.gpkg", driver="GPKG")
```
